## functions

```python
#we can efine our own functions
def my_print(obj): #obj is a parameter passed into this fn
    print(obj)
my_print("hello world") #calling/invoking the fn
```

```python
#functions do work and can return something
def multiply_by_7(num):
    #after everything is done above, return some value
    return_val = num*7
    return return_val
#my_num stores the returned value from caling the function and passing some parameter
my_num = multiply_by_7(3)
my_num
```

```python
#what does the print() return
print_return = print("codig is fun")
print(print_return)
type(print_return)
```

```python
#the nontype is the absence of value
print_return+1
```

## function parameters

```python
#a fn with multiple parameters
def combine_strings (str1, str2, separator):
    return str1 + separator + str2
combine_strings("hey", "guys", ' ')
```

```python
#what happens when i forget the last parametet
combine_strings("hello", "world")
```

```python
#make a default parameter value for separator
def combine_strings (str1, str2, separator=' '):
    return str1 + separator + str2 
combine_strings("niaje", "guys")
    
```

```python
#using keyword parameters (arguments)
def combine_strings(str1, str2, separator= ' ', message = ' '):
    if message != '':
        print(message)
        #below shows the order inwhich they will be returned
    return str1+ separator  + str2 
combine_strings("hello", "world", message ="coding is fun!!")
```

```python
#named parameters come after positional parameters
combine_strings("hello", str2="world", message ="coding is fun!!", separator=' ')
```

## variable scope

```python
#this variable has a global scope - can be accessed anywhere below
my_string= "my string"
print(my_string)
```

```python
my_string
```

```python
#functions can access global variables
def print_my_string():
     print(my_string)
print_my_string()
   
```

```python
#functions can ghave local variables- defined inside the fn

def print_my_string(my_string):
    print(my_string)
print_my_string("another string")
print(my_string)
```

```python
#functions can have local variables
def print_my_string():
    my_string = "This is a local string variable"
    print(my_string)
print_my_string()
print("global string variable:", my_string)
```

```python
#global variables can be changed by functions
my_list = [] #global variable
def append_list():
    my_list.append("stuff")
append_list()
my_list
```

```python
#changing a global variable inside a function
state = 200
def login():
    if state == 200:
        print("logged in successfully")
    else:
            print("try again")
login()
```

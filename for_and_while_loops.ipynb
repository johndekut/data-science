# for loops

```python
#used for controlling flow when code is to be executed more than once
#create a range object 
my_range = range(0, 10)
print(my_range)
print(list(my_range))
```

```python
#use the range eith a for loop 
#the f means formatted string lateral and it lets you insert variables inside {}
for value in my_range:
    if value ==2:
        print(f"i care about {value}")
    else:
       print(f"i dont care about {value}")
```

```python
#use a list with for loop
#'string' refers to a value inside the list, could be any word
string_list =["one", "two", "three", "four", "five"]
for string in string_list:
    print(string)
```

```python
#build a list using for loop
my_list = []
for number in my_range:
    #if the value is 2 then skip it
    if number ==2:
        continue #go back up and rerun the code
    my_list.append(number)
my_list
```

```python
#build a list with for loop
#break stops further executions
my_list = []
for value in my_range:
    #if the value is 5 stop
    if value ==5:
        break
    my_list.append(value)
my_list
```

## while loop

```python
#runs a block of code as long as a certain condition is true
#could be dangerous if there is no stopping critera
```

```python
a = True
while a:
    print("a is true)
```

```python
#better while example
a= 0
while a <10:
    print(a)
    a+=1 #increment stage
```

```python
#another dangerous example
a = 0
while a<20:
    if a==4:
        continue
    print(a)
    a+=1
```

```python
a= 0
while a <10:
    if a != 7:
        print(a)
    a=a+1
```

```python
#breaking out of a while loop
a = 0
while a <10:
    if a==7:
        break
        
    print(a)
    a+=1
    
    
```

## list comprehensions

```python
#its a more pythonic way of wring for loops
#build a new list from an exisiting list
score_list = [23, 56, 45, 76, 78, 89, 94]
#square every value in my score_list
new_list = [value **2 for value in score_list]
new_list
```

```python
#you can use filters
my_list = list(range(1, 101))
#only square the Nos in my_list if its less than 50
new_list = [number**2 for number in my_list if number <=50]
print(new_list)
```

```python
#you can use functions
new_list = [float(x) for x in my_list if x>50]
print(new_list)
```

```python
#list comprehensions in spatial analysis
coordinates = [(34.56, -0.12), (34.59, -0.15), (34.61, -0.20), (34.55, -0.10)]
from math import radians, sin, cos, sqrt, atan2
```

```python
base = (34.55, -0.10)
def formula(p1, p2):
    R = 6371
    lon1, lat1, lon2, lat2 = map(radians, [p1[0], p1[1], p2[0], p2[1]])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    return 2 * R *atan2(sqrt(a), sqrt(1-a))
#compute distances for all points
distances = [formula(base, p) for p in coordinates]
#filter points only within 5 km
nearby_points = [p for p in coordinates if formula(base, p)<=5]
print("Distances (km):", [round(d, 2) for d in distances])
print("Nearby points:", nearby_points)
```

### using lambdas

```python
#lambdas are small anonymous function
(lambda x: x =="stuff")(9) #9 will take the value of x
```

```python
#lambdas are functions and functions are objects
x = (lambda x: x == "stuff")
print(type(x))
x('stuff')
```

```python
#lambdas can use if-else
(lambda x: print(True) if x =="stuff" else print(False))("stuff")
```

```python
#lambdas can be passed into other functions
my_list= [10, 20, 30, 40, 50, 60, 70, 80]
new_list =list( filter(lambda val: val>20, my_list))
new_list
```

```python
students = [
    ("Jonte", 75),
    ("Margaret", 88),
    ("Brian", 62),
    ("Faith", 90),
    ("Kevin", 68)
]
#sort the marks by ascending order, highest >> lowest
sorted_students= sorted(students,key=lambda s: s[1], reverse= True)
top_students= list(map(lambda s: s[0], filter(lambda s:s[1]>70,sorted_students)))
print(top_students)

```
